## Impute missing values using the mean 

Fill the missing `'Age'` values using the average age. (Don't overwrite the original data, as we will be comparing to other methods for dealing with the missing values.) Then recalculate the mean, median, and std and replot the histogram.

```python
# Your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
titanic_data = pd.read_csv('titanic.csv')

# Create a copy of the original DataFrame
titanic_data_filled = titanic_data.copy()

# Calculate the mean age
mean_age = titanic_data['Age'].mean()

# Fill missing 'Age' values with the mean age
titanic_data_filled['Age'].fillna(mean_age, inplace=True)

# Recalculate mean, median, and standard deviation of the 'Age' feature
new_mean_age = titanic_data_filled['Age'].mean()
new_median_age = titanic_data_filled['Age'].median()
new_std_age = titanic_data_filled['Age'].std()

plt.figure(figsize=(10, 6))
plt.hist(titanic_data_filled['Age'], bins=30, color='blue', alpha=0.7,edgecolor = 'black')
plt.title('Distribution of Age (Missing Values Filled with Mean)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Commentary

Note that the standard deviation dropped, the median was slightly raised and the distribution has a larger mass near the center.

## Impute missing values using the median 

Fill the missing `'Age'` values, this time using the median age. (Again, don't overwrite the original data, as we will be comparing to other methods for dealing with the missing values.) Then recalculate the mean, median, and std and replot the histogram.

```python
# Your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
titanic_data = pd.read_csv('titanic.csv')

# Create a copy of the original DataFrame
titanic_data_filled_median = titanic_data.copy()

# Calculate the median age
median_age = titanic_data['Age'].median()

# Fill missing 'Age' values with the mean age
titanic_data_filled_median['Age'].fillna(median_age, inplace=True)

# Recalculate mean, median, and standard deviation of the 'Age' feature
new_mean_age_median = titanic_data_filled_median['Age'].mean()
new_median_age_median = titanic_data_filled_median['Age'].median()
new_std_age_median = titanic_data_filled_median['Age'].std()

plt.figure(figsize=(10, 6))
plt.hist(titanic_data_filled_median['Age'], bins=30, color='blue', alpha=0.7,edgecolor = 'black')
plt.title('Distribution of Age (Missing Values Filled with Median)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Commentary

Imputing the median has similar effectiveness to imputing the mean. The variance is reduced, while the mean is slightly lowered. You can once again see that there is a larger mass of data near the center of the distribution.
